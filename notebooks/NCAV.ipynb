```python
import pandas as pd
pd.set_option("display.precision", 1)
corp_info = pd.read_csv('corp.csv', header=0)
corp_info
```

```python
corp_info = corp_info[['종목코드','종목명','현재가','상장주식수']]
corp_info
```

```python
corp_info['종목코드'] = 'A' + corp_info['종목코드'].astype(str)
corp_info
```

```python
corp_result = pd.read_excel('SAMPLE_2018.xlsx')
corp_result
```

```python
corp_result.columns
```

```python
corp_data = corp_result[['종목코드', '유동자산', '부채총계']]
corp_data
```

```python
corp_total = pd.merge(corp_info, corp_data, how='left', on='종목코드')
corp_total
```

```python
corp_total = corp_info.join(corp_data.set_index('종목코드'), how='outer', on='종목코드')
corp_total
```

```python
corp_total_dropna = corp_total.dropna()
corp_total_dropna
```

```python
corp_total_dropna["상장주식수"] = corp_total_dropna["상장주식수"].str.replace(",","")
corp_total_dropna
```

```python
#corp_total_dropna["시가총액"] = pd.to_numeric(corp_total_dropna["상장주식수"]) * corp_total_dropna["현재가"]
corp_total_dropna["시가총액"] = corp_total_dropna["현재가"].str.replace(",","").astype('float64') * corp_total_dropna["상장주식수"].astype('float64')
corp_total_dropna
```

```python
corp_total_dropna["NCAV"] = corp_total_dropna["유동자산"] - corp_total_dropna["부채총계"]
corp_total_dropna["NCAV_RESULT"] = corp_total_dropna["NCAV"] > corp_total_dropna["시가총액"]*1.5
corp_total_dropna
```
